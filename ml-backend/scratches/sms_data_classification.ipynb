```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import LabelEncoder
from sklearn.svm import LinearSVC
from sklearn.model_selection import train_test_split
import numpy as np
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_excel("sms-data.xls", sheetname='Sheet1')
```

```python
df
```

```python
target = np.array(df['class'])
```

```python
vectorizer = TfidfVectorizer(ngram_range=(1, 3))
vectorizer.fit(df['Body'])

label_encoder = LabelEncoder()
label_encoder.fit(list(df['class']))
```

```python
dataset = vectorizer.transform(df['Body']).toarray()
```

```python
target = label_encoder.transform(target)
```

```python
X_train, X_test, y_train, y_test = train_test_split(dataset, target, test_size=0.50, random_state=42)
```

```python
model = LinearSVC()
model.fit(X_train,y_train)

pred = model.predict(X_test)
print("LinearSVC accuracy : ", accuracy_score(y_test, pred, normalize=True))
```

```python
pred = model.predict(dataset)
df['predicted'] = label_encoder.inverse_transform(pred)
```
