```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from ml_models.debit_credit_classifier import DClassifier
```

```python
df = pd.read_csv("../scratches/merge-data.csv")
df = df.fillna("NA")
en = df.head(10)
is_debit_or_credit =  df['class']!="NA"
df = df[is_debit_or_credit]
df= pd.concat([df,en])
dataset = np.array(df['body'],dtype=np.unicode_)
target = np.array(df['class'],dtype=np.unicode_)
print(df.tail())
print(dataset,target)
```

```python
model = DClassifier()
target = []
for i in dataset:
    pred = model.predict_label(i)
    target.append(pred[0])
df['label'] = target
```
